```python
#1. Loading Data:

import pandas as pd

# Load dataset from a URL or local CSV
url = "C:/Users/Netravati/Downloads/archive/WineQT.csv" 
df = pd.read_csv(url)

# Display the first few rows
print(df.head())

# Check the structure of the dataset
print(df.info())
print(df.describe())


```

```python
 # 2.Checking for Missing Values:

# Identify missing values
missing_values = df.isnull().sum()
print(missing_values)


```

```python
 #3. Data Filtering:

# Filter rows where alcohol > 10
high_sugar_wines = df[df['residual sugar'] >5]

# Filter rows where quality >= 7
high_quality_wines = df[df['quality'] >= 5]




```

```python
 #4. Sorting Data:

# Sort by quality (ascending)
df_sorted_quality = df.sort_values('quality')

# Sort by alcohol (descending)
df_sorted_alcohol = df.sort_values('free sulfur dioxide', ascending=False)
```

```python
#5. Group By and Aggregation:

# Group by quality and calculate mean
group_quality_mean = df.groupby('quality').mean()

# Calculate median alcohol by quality level
median_alcohol_by_quality = df.groupby('quality')['alcohol'].median()

```

```python
#6. Adding New Columns:

# Categorize wines into High and Low Quality
df['quality_category'] = df['quality'].apply(lambda x: 'High Quality' if x > 7 else 'Low Quality')

```

```python
#NumPy: Statistical Analysis

# 1. Basic Statistics:

import numpy as np

# Calculate mean, median, std, min, max for alcohol, pH, citric acid
acidity_mean = np.mean(df['fixed acidity'])
acidity_median = np.median(df['fixed acidity'])
acidity_std = np.std(df['fixed acidity'])
acidity_min = np.min(df['fixed acidity'])
acidity_max = np.max(df['fixed acidity'])

```

```python
#2. Correlations:

# Calculate correlation between alcohol and quality
alcohol_quality_corr = np.corrcoef(df['alcohol'], df['quality'])[0, 1]

# Correlation between residual sugar and quality
sugar_quality_corr = np.corrcoef(df['residual sugar'], df['quality'])[0, 1]

```

```python
#3. Working with Arrays:

# Convert alcohol column to NumPy array
alcohol_array = df['alcohol'].values

# Normalize the values
alcohol_normalized = (alcohol_array - alcohol_array.min()) / (alcohol_array.max() - alcohol_array.min())

# Find unique quality values
unique_quality = np.unique(df['quality'])

```

```python
#. Statistical Analysis (T-test):

from scipy import stats

# Compare mean alcohol content of High vs Low Quality wines
high_quality_wines = df[df['quality'] > 7]
low_quality_wines = df[df['quality'] <= 7]

t_stat, p_val = stats.ttest_ind(high_quality_wines['alcohol'], low_quality_wines['alcohol'])


```

```python
#Matplotlib: Data Visualization

#1. Histograms:

import matplotlib.pyplot as plt

# Plot a histogram for alcohol
plt.hist(df['alcohol'], bins=20)
plt.title('Alcohol Distribution')
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.show()
```

```python
#2. Scatter Plots:

# Scatter plot of alcohol vs quality
plt.scatter(df['alcohol'], df['quality'])
plt.title('Alcohol vs Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()
```

```python
#3. Box Plots:

# Box plot of residual sugar by quality
df.boxplot(column='residual sugar', by='quality')
plt.title('residual Sugar by Quality')
plt.xlabel('Quality')
plt.ylabel('residual Sugar')
plt.show()
```

```python
#4. Bar Plots:

# Bar plot of average citric acid by quality
avg_citric_acid_by_quality = df.groupby('quality')['citric acid'].mean()
avg_citric_acid_by_quality.plot(kind='bar')
plt.title('Average Citric Acid by Quality')
plt.xlabel('Quality')
plt.ylabel('Citric Acid')
plt.show()

```

```python
#5. Line Plots:

# Line graph of average pH level by quality
avg_pH_by_quality = df.groupby('quality')['pH'].mean()
avg_pH_by_quality.plot(kind='line')
plt.title('Average pH by Quality')
plt.xlabel('Quality')
plt.ylabel('pH')
plt.show()


```
